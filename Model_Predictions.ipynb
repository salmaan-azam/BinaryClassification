```python
import pandas as pd
data = pd.read_csv("training_set.csv", index_col = 0)
y_train = data.Y                                                
X_train = data.drop(['Y'], axis=1)
X_test  = pd.read_csv("test_set.csv", index_col = 0)  
```

```python
print(X_train.shape)
print(X_test.shape)
print(len(y_train))
```

```python
def test_result(X_train, X_test, y_train):
  
  """This function is for getting test result"""
  import xgboost as xgb
  import numpy as np
  from sklearn.metrics import roc_auc_score, accuracy_score
  from sklearn.preprocessing import StandardScaler
  
  ss = StandardScaler()
  X_train = ss.fit_transform(X_train)
  X_test = ss.transform(X_test)

  clf1 = xgb.sklearn.XGBClassifier(n_estimators=500, max_depth=3, learning_rate=0.1,  verbosity=1, objective='binary:logistic', n_jobs=-1)
  clf1.fit(X_train, y_train)

  y_train_pred = clf1.predict(X_train)
  y_train_prob = clf1.predict_proba(X_train)

  print(f'AUC for train data : {np.round(roc_auc_score(y_train, y_train_prob[:,1]),4)}')
  print(f'Accuracy Value for train data : {np.round(accuracy_score(y_train, y_train_pred)*100,2)}%')
  
  y_test_pred = clf1.predict(X_test)
  y_test_prob = clf1.predict_proba(X_test)

  return y_test_pred, y_test_prob
```

```python
y_test_pred, y_test_prob = test_result(X_train, X_test, y_train)
```

```python
print(len(y_test_pred), len(y_test_prob))
```
